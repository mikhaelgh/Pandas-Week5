```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
#import modul
```

```python
n_rows = 365 * 24
n_cols = 2
cols = ['col1', 'col2']

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                 columns=cols)

df.index = pd.util.testing.makeDateIndex(n_rows, freq='H')
df

#persiapan data frame, kolom baris diisi 365 * 24, 24 adalah 24 jam sehari, 365 adalah jumlah hari setahun
```

```python
df.resample('M')['col1'].sum().to_frame()
#interval waktu berdasarkan bulan dengan
```

```python
df.resample('D')['col1'].sum().to_frame()
#interval waktu berdasarkan hari  
```
